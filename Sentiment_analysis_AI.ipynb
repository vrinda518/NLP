### LOADING THE DATASET

```python
import pandas as pd
df = pd.read_csv("friends.csv")
print(df.head())
```

```python
labels = []
for sent in df['Sentiment']:
  if sent == 'NEGATIVE':
    labels.append(0.0)
  elif sent == 'POSTIVE':
    labels.append(1.0)
  else:
    labels.append(2.0)
df["labels"] = labels
```

```python
df.head()
y = df.iloc[:,-1]
```

### DATA PREPROCESSING

```python
index_names=df[df['Sentiment']=='MIXED'].index
df=df.drop(index_names)
```

```python
df['Sentiment'].unique()
```

```python
import nltk
nltk.download('punkt')
```

```python
import re
def  clean_text(df, text_field, new_text_field_name):
    df[new_text_field_name] = df[text_field].str.lower()
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|^rt|http.+?", "", elem))  
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: re.sub(r"\d+", "", elem)) 
    return df
data_clean = clean_text(df, 'Scene', 'Scene_Clean')
data_clean.head()
```

```python
from nltk import word_tokenize
clean_tokens = [word_tokenize(sent) for sent in df['Scene_Clean']]
```

```python
print(clean_tokens)
```

```python
from nltk import PorterStemmer
porter = PorterStemmer()
stemmed_tokens = [[porter.stem(token) for token in l] for l in clean_tokens]
print(stemmed_tokens)
```

```python
stemmed_list = []
for l in stemmed_tokens:
  stemmed_list.append(' '.join(l))
print(stemmed_list)
```

```python
df["cleaned"] = stemmed_list
```

```python
print(df.head())
```

### COUNT VECTORIZER

```python
from sklearn.feature_extraction.text import CountVectorizer,ENGLISH_STOP_WORDS
stop = ENGLISH_STOP_WORDS.union(['phoebe','monica','chandler','rachel','joey','ross','rach'])
cv = CountVectorizer(ngram_range = (1,2),stop_words = stop,max_features=5000)
cv.fit_transform(df['cleaned'])
bow = cv.transform(df['cleaned'])
print(bow.toarray())
X_df = pd.DataFrame(bow.toarray(), columns=cv.get_feature_names())
print(X_df.head())
##print(X_df['rach'])
```

```python
y = df.iloc[:,2]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_df,y,test_size=0.2,random_state=92)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(penalty='l2',C = 1)
log_reg = lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = log_reg.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

### TfIdf

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(stop_words = stop)
vect = tf.fit(df['cleaned'])
bowtf = tf.transform(df['cleaned'])
print(bowtf.toarray())
X_tf = pd.DataFrame(bowtf.toarray(), columns=tf.get_feature_names())
print(X_tf)
```

```python
X_train2,X_test2,y_train2,y_test2 = train_test_split(X_tf,y,test_size=0.2,random_state=92)
```

```python
log_reg2 = lr.fit(X_train2,y_train2)
y_pred2 = log_reg2.predict(X_test2)
print(confusion_matrix(y_test2,y_pred2))
print(accuracy_score(y_test2,y_pred2))
```
